## Science des données
</div>

Colab vous permet de tirer pleinement parti des bibliothèques populaires Python pour analyser et visualiser des données. La cellule de code ci-dessous utilise <strong>numpy</strong> pour générer des données aléatoires et <strong>matplotlib</strong> pour les visualiser. Pour modifier le code, cliquez simplement sur la cellule.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
#stratified sampling
from sklearn.model_selection import train_test_split

# Assuming you want to stratify based on 'Value',
# Create bins for 'Value' to categorize it.
df['Category'] = pd.cut(df['Value'], bins=5, labels=False)

# Now perform stratified sampling using the new 'Category' column
train, stratified_sampled_data = train_test_split(df, test_size=0.2, stratify=df['Category'], random_state=42)
print(stratified_sampled_data)
```

```
# Over-Sampling: Adaptations
import numpy as np
df['Class'] = np.random.choice(['minority', 'majority'], size=100, p=[0.05, 0.95])
print(df)
```

```
#Over-Sampling: Application
from sklearn.utils import resample
minority_class= df [df [ 'Class'] == 'minority']
majority_class = df [df ['Class'] == 'majority']
print(minority_class)
oversampled_minority = resample(minority_class, replace=True, n_samples=500, random_state=42)
oversampled_sample =pd.concat([majority_class, oversampled_minority])
```

```
#Undersampling
minority_class = df [df ['Class'] == 'minority']
majority_class= df [df ['Class'] == 'majority']
undersampled_majority =resample(majority_class, replace=False, n_samples=len(minority_class), random_state=42)
undersampled_sample =pd.concat([undersampled_majority, minority_class])
```

### Manual Definition

```
data={
    'City':['Casablance','Mohammedia','Agadir','Taghazout','Errachida','Midelt'],
    'Region':['Grand Casablanca','Grand Casablanca','Souss-Massa','Souss-Massa','Daraa-Tafilalt','Daraa-Tafilalt'],
    'Country':['Morocco','Morocco','Morocco','Morocco','Morocco','Morocco']
}
df=pd.DataFrame(data)
print(df)
```

### binning-based Hierarchy

```
data={'Income':[1000,2500,4000,7000,12000]}
df=pd.DataFrame(data)
bins=[0, 3000, 7000, float('inf')]#99999
labels=['low','medium','high']
df['income_level']=pd.cut(df['Income'],bins=bins,labels=labels)
print(df)
```

### Top-down-segmentation

```
data={
    'City':['Casablance','Mohammedia','Agadir','Taghazout','Errachida','Midelt'],
    'Region':['Grand Casablanca','Grand Casablanca','Souss-Massa','Souss-Massa','Daraa-Tafilalt','Daraa-Tafilalt'],
    'Country':['Morocco','Morocco','Morocco','Morocco','Morocco','Morocco']
}
df=pd.DataFrame(data)
print(df)
```

### clustring-based hierarchy

```
data={'Customer':['C1','C2','C3','C4','C5'], 'Spending Amount':[1000,2500,4000,7000,12000]}
df=pd.DataFrame(data)
kmeans= KMeans(n_clusters=3, random_state=None)
df['segment'] = kmeans.fit_predict(df[['Spending Amount']])
cluster_centers = kmeans.cluster_centers_.flatten()  # Corrected line
sorted_centers=sorted(range(len(cluster_centers)),key=lambda x: cluster_centers[x])
df['segment Label']=df['segment'].map({sorted_centers[0]:'low',1:'medium',2:'high'})
print(df)

```
